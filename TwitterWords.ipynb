```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
import matplotlib.pyplot as plt
import networkx as nx

tweets = [
    "Machine learning is amazing.",
    "Data science involves machine learning.",
    "Natural language processing is part of machine learning."
]

#Find word pairs
word_pairs = []
for sentence in tweets:
    words = word_tokenize(sentence) #Cut the sentence into words
    pairs = ngrams(words, 2) #Find pairs of words
    word_pairs.extend(pairs) #Add the pairs to our list
    
# Make a word network
word_network = nx.Graph()
for word1, word2 in word_pairs:
    word_network.add_edge(word1, word2)

#Draw the word network
nx.draw(word_network, with_labels=True)
plt.show()
```
